```python
#get rid of null values in positions, as we will sort the datasets based on position

fifa_lst = [fifa_2016, fifa_2017, fifa_2018, fifa_2019, fifa_2020]

for fifa in fifa_lst:
    fifa = fifa.dropna(subset = ["team_position", "player_positions"])
    
fifa_2016 = fifa_2016[fifa_2016["value_eur"] != 0]
fifa_2017 = fifa_2017[fifa_2017["value_eur"] != 0]
fifa_2018 = fifa_2018[fifa_2018["value_eur"] != 0]
fifa_2019 = fifa_2019[fifa_2019["value_eur"] != 0]
fifa_2020 = fifa_2020[fifa_2020["value_eur"] != 0]
```

```python
#sort the datasets by specific position into position groupings

fifa_lst = [fifa_2016, fifa_2017, fifa_2018, fifa_2019, fifa_2020]

forward_group_lst = ["ST", "CF", "RF", "LF", "RS", "LS"]
winger_group_lst = ["RW", "LW", "RM", "LM"] 
attmid_group_lst = ["CAM", "RAM", "LAM"]
midfield_group_lst = ["CM", "CDM", "RCM", "LCM", "RDM", "LDM"]
outback_group_lst = ["LB", "RB", "LWB", "RWB"]
cenback_group_lst = ["CB", "LCB", "RCB"]

for fifa in fifa_lst:
    group_lst = []
    for index, row in fifa.iterrows():
        if row["team_position"] in forward_group_lst:
            group_lst.append("forward")
        elif row["team_position"] in winger_group_lst:
            group_lst.append("winger")
        elif row["team_position"] in attmid_group_lst:
            group_lst.append("attmid")
        elif row["team_position"] in midfield_group_lst:
            group_lst.append("midfield")
        elif row["team_position"] in outback_group_lst:
            group_lst.append("outback")
        elif row["team_position"] in cenback_group_lst:
            group_lst.append("cenback")
        elif row["team_position"] == "GK":
            group_lst.append("goalkeeper")
        else:
            potential_group_lst = row["player_positions"].split(",")
            if potential_group_lst[0] in forward_group_lst:
                group_lst.append("forward")
            elif potential_group_lst[0] in winger_group_lst:
                group_lst.append("winger")
            elif potential_group_lst[0] in attmid_group_lst:
                group_lst.append("attmid")
            elif potential_group_lst[0] in midfield_group_lst:
                group_lst.append("midfield")
            elif potential_group_lst[0] in outback_group_lst:
                group_lst.append("outback")
            elif potential_group_lst[0] in cenback_group_lst:
                group_lst.append("cenback")
            elif potential_group_lst[0] == "GK":
                group_lst.append("goalkeeper")
    fifa["position_group"] = group_lst
```

```python
FIFA2020_FINAL_dict = FIFA2020_FINAL.to_dict(orient = 'list')
for key, value in FIFA2020_FINAL.items():
    valuation_lst = []
    if key == "value_eur":
        value_lst = value
    elif key == "prediction":
        preds_lst = value

        
for i in range(len(value_lst)):
    if value_lst[i] < preds_lst[i]:
        valuation_lst.append("undervalued")
    if value_lst[i] > preds_lst[i]:
        valuation_lst.append("overvalued")
    if value_lst[i] == preds_lst[i]:
        valuation_lst.append("perfect")
            
FIFA2020_FINAL["valuation"] = valuation_lst

        


```

```python
def playerValuationSearch(df, player):
    return(df[df["short_name"] == player][["short_name", "value_eur", "prediction", "valuation"]])


print(playerValuationSearch(FIFA2020_FINAL, "Cristiano Ronaldo"))
```
